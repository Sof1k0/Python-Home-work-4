```python
purchases = {}
```

```python
with open("purchase_log.txt", "r", encoding='UTF-8') as purchases_file:
    for line in purchases_file:
        id, name = line.split(" ")
        name = name.replace('\n','')
        name = name.strip("`")
        # print(f"Id : {id}, Name : {name}")
        
        if id and name:
            purchases[id] = name
            
print(purchases)
```
