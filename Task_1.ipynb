```python
#1. перечень всех документов*

documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

#2. перечень полок, на которых хранятся документы (если документ есть в documents, он обязательно должен быть и в directories)*

directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}
```

```python
def doc_number():
    num = input("Введите номер документа:")
    for dic in documents:
        if "number" in dic.keys() and dic["number"] == num:
            print("Владелец документа: "+ dic["name"])
            break
    else:
        print("Документ не найден в базе")
```

```python
def doc_shelf():
    numb = input("Ввведите номер документа")
    list_numb = []
    for i in directories.values():
        for ii in i:
            list_numb.append(ii)
    if numb in list_numb:
        for item, values in directories.items():
            if numb in values:
                result = "Документ хранится на полке: " + item
    else:
        result = "Документ не найден в базе"
    print(result)
```

```python
def main():
    inp_dict = {'p' : doc_number, 's' : doc_shelf, 'q' : False}
    y = True
    while y:
        choice = input('Введите команду').lower()
        if choice in inp_dict.keys():
            if choice == 'q':
                y = False
            else:
                inp_dict[choice]()
        else:
            print('Такой команды нет')
```

```python
main()
```
